```python
import matplotlib.pyplot as plt
```

```python
# function y = (x + 3)^2
def func(x):
    return (x + 3)**2

# derivative (gradient)
def grad(x):
    return 2 * (x + 3)
```

```python
# initialize variables
x = 2.0            # starting point
alpha = 0.1        # learning rate
iters = 20         # number of iterations

# for plotting
x_values = []
y_values = []

print("Iteration\tX Value\t\tY Value")
for i in range(iters):
    y = func(x)
    print(f"{i+1}\t\t{x:.5f}\t\t{y:.5f}")
    x_values.append(x)
    y_values.append(y)
    
    # update step
    x = x - alpha * grad(x)

# final values
print("\nLocal Minima at x =", round(x, 5))
print("Minimum Value of y =", round(func(x), 5))
```

```python
plt.plot(range(1, iters+1), y_values, marker='o')
plt.title("Gradient Descent Convergence")
plt.xlabel("Iteration")
plt.ylabel("Function Value (y)")
plt.grid(True)
plt.show()
```
